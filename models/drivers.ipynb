```python
import pandas as pd

data = pd.read_csv("dl_data.csv")

data["Gender"] = data["Gender"].apply(lambda x: int(x == "Male"))
data["Age Group"] = data["Age Group"].apply(lambda x: 0 if x == "Teenager" else 1 if x == "Young Adult" else 2)
data["Race"] = data["Race"].apply(lambda x: 0 if x == "White" else 1 if x == "Black" else 2)
data["Training"] = data["Training"].apply(lambda x: 1 if x == "Basic" else 2 if x == "Advanced" else 0)
data["Reactions"] = data["Reactions"].apply(lambda x: 1 if x == "Average" else 2 if x == "Fast" else 0)
data["Qualified"] = data["Qualified"].apply(lambda x: int(x == "Yes"))

features = data.columns.tolist()
features.remove("Applicant ID")
features.remove("Qualified")

print(features)

X = data[features]
y = data.Qualified

from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=0)

data.head()
```

```python
data.describe()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

forest_model = RandomForestClassifier(random_state=1)
forest_model.fit(train_X, train_y)

predictions = forest_model.predict(test_X)
print(classification_report(test_y, predictions))
```

```python
from pickle import dump

with open("forest_model.pickle", "wb") as f:
    dump(forest_model, f, protocol=5)
```
